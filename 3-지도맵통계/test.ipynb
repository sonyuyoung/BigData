# 다운받은 엑셀 파일에서, 1차 지역명, 2차 지역명 같은 컬럼에 있어서 -> 분리 작업

```python
import pandas as pd 

CB = pd.read_csv("./CoffeeBean.csv", encoding="cp949",index_col=0, header=0, engine="python")
# 메모리상에올려서 기본 상위5개 
CB.head()
```

```python
# 행정구역별 시군구 인구수 자료에서 , -> 현재 필요한 부분은 지역 컬럼이 필요함. 정리하기
# 임시로 담을 리스트 
addr=[]
for address in CB.address: 
    # 서울시 강남구 학동로 211 1층 구분자 " "공백 
    # 구분자(공백) 를 기준으로 나누기 
    addr.append(str(address).split())
addr
```

```python
# 1차 지역명 정리 작업 : 예 ) 서울 , 서울시 , 서울특별시 변경작업 
addr2 =[]

for i in range(len(addr)):
    if addr[i][0]=="서울":addr[i][0]="서울특별시"
    elif addr[i][0]=="서울시":addr[i][0]="서울특별시"
    elif addr[i][0]=="부산시":addr[i][0]="부산광역시"
    elif addr[i][0]=="인천":addr[i][0]="인천광역시"
    elif addr[i][0]=="광주":addr[i][0]="광주광역시"
    elif addr[i][0]=="대전":addr[i][0]="대전광역시"
    elif addr[i][0]=="울산":addr[i][0]="울산광역시"
    elif addr[i][0]=="세종":addr[i][0]="세종특별자치시"
    elif addr[i][0]=="경기":addr[i][0]="경기도"
    elif addr[i][0]=="충북":addr[i][0]="충청북도"
    elif addr[i][0]=="충남":addr[i][0]="충청남도"
    elif addr[i][0]=="전북":addr[i][0]="전라북도"
    elif addr[i][0]=="전남":addr[i][0]="전라남도"
    elif addr[i][0]=="경북":addr[i][0]="경상북도"
    elif addr[i][0]=="경남":addr[i][0]="경상남도"
    elif addr[i][0]=="제주":addr[i][0]="제주특별자치도"
    elif addr[i][0]=="제주도":addr[i][0]="제주특별자치도"
    elif addr[i][0]=="제주시":addr[i][0]="제주특별자치도"
    elif addr[i][0]=="서귀포시":addr[i][0]="제주특별자치도"
    addr2.append(" ".join(addr[i]))
    addr2

# addr2.to_csv("./CoffeeBean_3.csv", encoding="cp949", index=False)
    

```

```python
# 데이터 프래임 표 형식에 넣어서 작업준비
# 판다스
addr2 = pd.DataFrame(addr2,columns=["address"])

# 기존에 CB(커피빈데이터) 병합 2개의 데이터프레임을 위아래로(수직으로) 합치는 방식
CB = pd.concat([CB,addr2],axis=1)
CB.head()
```

```python
# 데이터 프레임(메모리상에서 표) -> csv 파일 형태로 변환 작업. 
CB.to_csv("./CoffeeBean_2.csv", encoding="cp949", index=False)
```
